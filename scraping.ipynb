```python
import ssl

import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq

```

```python
imdb_url = 'https://www.shopclues.com/mens-clothing-t-shirts.html'
uclient = uReq(imdb_url)   #opening thr url
imdb_page = uclient.read()
uclient.close()
imdb_html = bs(imdb_page, 'html.parser')
show_list = imdb_html.find_all('div', {"class": 'row'})
```

```python
reviews = []
for show in show_list:
    for i in show.find_all('div' , {"class" : "column col3"}):
        title = i.span.text
        p_code = i.a["pid"]
        price = i.find_all('span' , {"class" : "p_price"})[0].text[1:]
        p_discount = i.find_all('span' , {"class" : "prd_discount"})[0].text
        reviewpage = "https://www.shopclues.com/search?q=" + p_code
        reviewRes = requests.get(reviewpage)
        reviewRes.encoding = 'utf-8'
        showreview_html = bs(reviewRes.text, "html.parser")
        pro = showreview_html.find("div" , {"class" : "column col3 search_blocks"})
        product_link = "https://"+ pro.a["href"][2:]
        mydict = {"title" : title ,"price" : price,"p_discount":p_discount,"p_code":p_code,"product_link":product_link}
        reviews.append(mydict)
        
        
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(reviews)
```
